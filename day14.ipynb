```python
#Q21
class laptop:
        def __init__(self, company, model, ram, rom, storage, screen_size, colour, cost):
            self.company = company
            self.model = model 
            self.ram = ram
            self.rom = rom
            self.storage = storage
            self.screen_size = screen_size
            self.colour = colour
            self.cost = cost
            
        def get_ram(self):
            return self.ram
dell=laptop("dell", "latitude", "6gb", "5gb", "1tb", "14inch", "black", "50k")
dell.get_ram()
```

```python
#Q22
class ATM:
    def __init__(self, bank, location, balance, card, pin):
        self.bank = bank
        self.location = location
        self.balance = balance
        self.card = card
        self.pin = pin
        
    def get_location(self):
        return self.location
sbi=ATM("sbi","pune", "0", "debit", "not to share")
sbi.get_location()
```
